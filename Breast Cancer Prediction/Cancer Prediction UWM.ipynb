```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import time
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score 
from sklearn.svm import SVC,LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
df =  pd.read_csv('Data.csv')
df = df.set_index('id')
df = df.iloc[:,:-1]
y = df.loc[:,['diagnosis']]
x = df.iloc[:,1:]
output_dict = {'M':1,'B':0}
print(y.head())
```

```python
y['diagnosis'] = y['diagnosis'].map(lambda x : output_dict.get(x))
print(y[:10])
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.2, random_state = 42)
print(xtrain.shape)
print(ytrain.head())
accuracies  = [] # list to store accuracy for n number of algorithms 
cvs = [] # cross-validartion score 
cost = [] # time is the measure of cost in this case
```

```python
model_algorithms=[LogisticRegression(),SVC(),LinearSVC(),DecisionTreeRegressor(),RandomForestClassifier(n_estimators=12),KNeighborsClassifier(n_neighbors=6)]
```

```python
for clf in model_algorithms:
    start_time = time.time()
    clf.fit(xtrain,ytrain)
    p = clf.predict(xtest)
    stop_time = time.time()
    print(p[:10])
    print(ytest[:10])
    prediction_vs_correct = pd.DataFrame({'P':p,'R':ytest.diagnosis})
    print(prediction_vs_correct[:10])
    time_taken = start_time - stop_time
    cost.append(time_taken)
    current_accuracy = accuracy_score(p,ytest.diagnosis)
    accuracies.append(current_accuracy)
    print("Current Accuracy:")
    print(current_accuracy)
    # cv = 5 -> Repeat the experiment 5 times
    current_cvs = cross_val_score(clf,x,y,cv=5)
    print(current_cvs)
    cvs.append(np.mean(current_cvs))
    print(cvs)
```

```python
# Random Forests Give the best Accuracy For Cancer Prediction with accuracy = 97%
```
